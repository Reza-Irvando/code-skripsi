```python
import os
import rasterio
import numpy as np
import joblib
from rasterio.windows import Window
from sklearn.metrics import confusion_matrix, accuracy_score, cohen_kappa_score
from tqdm import tqdm # Import pustaka progres bar

# ================= 1. KONFIGURASI =================
DIR_ORTO = "eksport-orto"
DIR_CLASS = "hasil-ndvi-class"
tanggal_uji = "20231005-full"
NAMA_FILE_MODEL = "model_svm_terlatih.pkl"
CHUNK_SIZE = 100000 # Jumlah piksel yang diprediksi per satu langkah progres
```

```python
# ================= 2. MEMUAT MODEL =================
model_svm = joblib.load(NAMA_FILE_MODEL)
```

```python
# ================= 3. MEMBACA CITRA UJI =================
print(f"Membuka citra uji: {tanggal_uji}...")
path_orto_uji = os.path.join(DIR_ORTO, f"{tanggal_uji}.tif")
path_class_uji = os.path.join(DIR_CLASS, f"class-{tanggal_uji}.tif")

with rasterio.open(path_orto_uji) as src_orto_uji, rasterio.open(path_class_uji) as src_class_uji:
    common_rows = min(src_orto_uji.height, src_class_uji.height)
    common_cols = min(src_orto_uji.width, src_class_uji.width)
    win = Window(0, 0, common_cols, common_rows)
    
    meta_hasil = src_class_uji.meta.copy()
    meta_hasil.update(height=common_rows, width=common_cols)
    
    img_orto_uji = src_orto_uji.read(window=win)
    img_class_uji = src_class_uji.read(1, window=win)
    
    bands, rows, cols = img_orto_uji.shape
    orto_uji_flat = img_orto_uji.reshape(bands, rows * cols).T
    class_uji_flat = img_class_uji.flatten()
    
    # MASKING
    mask_valid = (class_uji_flat == 1) | (class_uji_flat == 2) | (class_uji_flat == 3)
    indices_valid = np.where(mask_valid)[0] # Ambil indeks piksel yang valid
    X_test_valid = orto_uji_flat[mask_valid]
    y_test_valid = class_uji_flat[mask_valid] 
```

```python
# ================= 4. PREDIKSI DENGAN PROGRES =================
print(f"Mengeksekusi prediksi (Total piksel: {len(X_test_valid)})...")

# Inisialisasi array untuk hasil prediksi
y_pred_valid = np.zeros(len(X_test_valid), dtype=np.uint8)

# Membagi prediksi menjadi beberapa bagian (chunks)
for i in tqdm(range(0, len(X_test_valid), CHUNK_SIZE), desc="Prediksi Spasial"):
    end = i + CHUNK_SIZE
    # Prediksi per bagian
    y_pred_valid[i:end] = model_svm.predict(X_test_valid[i:end])

# Rekonstruksi ke peta
hasil_flat = np.zeros(rows * cols, dtype=np.uint8)
hasil_flat[mask_valid] = y_pred_valid
peta_hasil_svm = hasil_flat.reshape(rows, cols)
```

```python
# ================= 5. UJI AKURASI & SIMPAN =================
print("\n--- HASIL EVALUASI ---")
cm = confusion_matrix(y_test_valid, y_pred_valid, labels=[1, 2, 3])
print(f"Overall Accuracy : {accuracy_score(y_test_valid, y_pred_valid)*100:.2f}%")
print(f"Kappa            : {cohen_kappa_score(y_test_valid, y_pred_valid):.4f}")
print("Confusion Matrix:\n", cm)

nama_file_hasil = f"hasil_svm_{tanggal_uji}.tif"
with rasterio.open(nama_file_hasil, 'w', **meta_hasil) as dst:
    dst.write(peta_hasil_svm, 1)

print(f"\nSelesai! File disimpan sebagai {nama_file_hasil}")
```
